```python
print("CALCULADORA BÁSICA")

print(end="\n")

print("As operações básicas:")

print(end="\n")

print("1: Adição")
print("2: Subtração")
print("3: Divisão")
print("4: Multiplicação")
print("5: Exponenciação") 
 
print(end="\n")

a = input("Escolha um cálculo matemático para realizar !")

print(end="\n")

while a not in ("1", "2", "3", "4", "5"):
     print("Escolha um número de 0 a 5.")
     a = input("Escolha um cálculo matemático para realizar !")

print(end="\n")


if a == "1":
     print ("O cálculo escolhido foi a adição (+). ")
elif a == "2":
     print("O cálculo escolhido foi a subtração (-). ")
elif a == "3": 
     print("O cálculo escolhido foi a divisão (/). ")
elif a == "4":
     print("O cálculo escolhido foi a multiplicação (*). ")
elif a == "5":
     print("O cálculo escolhido foi a exponenciação (**). ")
else:
    print("Escolha um número de 0 a 5.")

print(end="\n")

v1 = float(input("Qual o primeiro valor ?"))
print(f'O primeiro valor é {v1}.')

print(end="\n")

v2 = float(input("Qual o segundo valor ?"))
print(f'O segundo valor é {v2}.')

print(end="\n")

adição = (v1 + v2)
subtração = (v1 - v2)
divisão = (v1 / v2)
multiplicação = (v1 * v2)
exponenciação = (v1 ** v2)

print(end="\n")

if a == "1":
     print(f'Com os valores sendo {v1} e {v2}, temos como resultado do cálculo: {adição}.')
elif a == "2":
     print(f'Com os valores sendo {v1} e {v2}, temos como resultado do cálculo: {subtração}.')
elif a == "3":
     print(f'Com os valores sendo {v1} e {v2}, temos como resultado do cálculo: {divisão}.')  
elif a == "4" :
     print(f'Com os valores sendo {v1} e {v2}, temos como resultado do cálculo: {multiplicação}.') 
else :
     print(f'Com os valores sendo {v1} e {v2}, temos como resultado do cálculo: {exponenciação}.') 

print(end="\n")

print (" Até mais !")






```
